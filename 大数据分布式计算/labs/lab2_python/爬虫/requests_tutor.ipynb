```python
import requests
```

```python
# 发送GET请求获取到响应
r = requests.get('http://www.baidu.com')
# 响应状态码
r.status_code
```

```python
# 响应头
r.headers
```

```python
# 猜测的编码
r.encoding
```

```python
# 响应内容的字符串形式（用猜测出的编码去解析二进制内容）
r.text
# 显然编码猜错了
```

```python
# r.content 是响应内容的二进制形式，解码得到正确的字符串
r.content.decode('utf-8')
```

```python
# 也可以把猜错的字符串改成正确的
r.encoding='utf-8'
r.text
```

```python
import urllib.parse as urlparse
from urllib.parse import parse_qs
url='http://www.szse.cn/api/report/ShowReport/data?SHOWTYPE=JSON&CATALOGID=1815_stock&TABKEY=tab1&txtDMorJC=000001&txtBeginDate=2020-10-08&txtEndDate=2020-10-09&radioClass=00,20,30&txtSite=all&random=0.963547861157542'
parsed = urlparse.urlparse(url)
params = parse_qs(parsed.query)
url = 'http://www.szse.cn/api/report/ShowReport/data'
```

```python
r = requests.get(url, params=params)
r.status_code
```

```python
r.text
```

```python
import json
d = json.loads(r.text)
d
```

```python
d[0]['data'][0]
```
